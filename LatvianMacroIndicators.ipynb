```python
import pandas as pd
from matplotlib import pyplot as plt
import eurostat
import numpy as np
```

Data is from [Eurostat](https://ec.europa.eu/eurostat/databrowser/view/tec00001/default/table?lang=en)

```python
my_filter_pars = {'geo': ['LV'], 'unit' : ['CP_MEUR']}

data = eurostat.get_data_df('tec00001', filter_pars=my_filter_pars)

df_long = data.melt(id_vars=[('freq'), ('na_item'), ('unit'), ('geo\\TIME_PERIOD')], var_name='year', value_name='value')

print(df_long)
```

```python
plt.plot(df_long['year'], df_long['value'], label = "", marker = 'o', color = '#9E3039') # '#9E3039' latvian red colour
plt.title(',    . ,  2011 - 2022.')
plt.xlabel('')
plt.ylabel(' \n(.)')
plt.legend()
plt.grid(True)
```

Data for population from [Eurostat](https://ec.europa.eu/eurostat/databrowser/view/tps00001/default/table?lang=en)

```python
my_filter_pars = {'geo': ['LV']}

dataPop = eurostat.get_data_df('tps00001', filter_pars=my_filter_pars)

df_longPop = dataPop.melt(id_vars=[('freq'), ('indic_de'),('geo\TIME_PERIOD')], var_name='year', value_name='value')

df_longPop['Pop1000'] = df_longPop['value'] / 1000

df_longPop
```

```python
plt.plot(df_longPop['year'], df_longPop['value'], label = "", marker = 'o', color = '#9E3039') # '#9E3039' latvian red colour
plt.title(',   . ,   2012 - 2023.')
plt.xlabel('')
plt.ylabel('\n(.)')
plt.legend()
plt.grid(True)
```

Data from [Eurostat](https://ec.europa.eu/eurostat/databrowser/view/tec00114/default/table?lang=en)

```python
my_filter_pars = {'geo': ['LV', 'BG']}

dataPop = eurostat.get_data_df('tec00114', filter_pars=my_filter_pars)

df_longPC = dataPop.melt(id_vars=[('freq'), ('na_item'), ('ppp_cat'), ('geo\TIME_PERIOD')], var_name='year', value_name='value')

df_longPC
```

```python
for country in df_longPC['geo\\TIME_PERIOD'].unique():
    country_data = df_longPC[df_longPC['geo\\TIME_PERIOD'] == country]
    if country == 'LV':
        line_color = '#9E3039'
    elif country == 'BG':
        line_color = '#00966E'
    plt.plot(country_data['year'], country_data['value'], label=country, marker='o', linestyle='-', markersize=8, color=line_color)
plt.axhline(y = 100, label = "EU")
plt.title('         2012 - 2022.')
plt.xlabel('')
plt.legend()
plt.grid(True)
```

```python
# Data for Imports of goods and services in % of GDP
my_filter_pars = {'geo': ['LV', 'BG']}

dataImp = eurostat.get_data_df('tet00004', filter_pars=my_filter_pars)

df_longImp = dataImp.melt(id_vars=[('freq'), ('na_item'), ('unit'), ('geo\TIME_PERIOD')], var_name='year', value_name='valueImp')

df_longImp
```

```python
# Data for Exports of goods and services in % of GDP
my_filter_pars = {'geo': ['LV', 'BG']}

dataExp = eurostat.get_data_df('tet00003', filter_pars=my_filter_pars)

df_longExp = dataExp.melt(id_vars=[('freq'), ('na_item'), ('unit'), ('geo\TIME_PERIOD')], var_name='year', value_name='valueExp')

df_longExp
```

```python
# New dataframe adding imports and exports as % ot GDP
TradeOpen = pd.merge(df_longImp[['geo\TIME_PERIOD', 'year', 'valueImp']], df_longExp[['geo\TIME_PERIOD', 'year', 'valueExp']])
TradeOpen['TradeOpen'] = TradeOpen['valueImp'] + TradeOpen['valueExp']
TradeOpen
```

```python
for country in TradeOpen['geo\\TIME_PERIOD'].unique():
    country_data = TradeOpen[df_longPC['geo\\TIME_PERIOD'] == country]
    if country == 'LV':
        line_color = '#9E3039'
    elif country == 'BG':
        line_color = '#00966E'
    plt.plot(country_data['year'], country_data['TradeOpen'], label=country, marker='o', linestyle='-', markersize=8, color=line_color)
plt.title('         2011-2022.')
plt.xlabel('')
plt.ylabel(' +   %  ')
plt.legend()
plt.grid(True)
```

Data from world bank: https://www.imf.org/external/datamapper/PPPPC@WEO/LVA?zoom=LVA&highlight=LVA

```python
gdppc = pd.read_excel("D:\latvia\imf-dm-export-20231226.xls")
gdppc.replace('no data', np.nan, inplace=True)
gdppc = gdppc.melt(id_vars=[('GDP per capita, current prices (Purchasing power parity; international dollars per capita)')], var_name='year', value_name='GDPPC, PPP')
gdppc = gdppc.dropna()
gdppc['GDPPC, growth'] = gdppc['GDPPC, PPP'].pct_change() * 100
gdppc['GDPPC_log'] = np.log(gdppc['GDPPC, PPP'])
gdppc
```

```python
average_growth = gdppc["GDPPC, growth"].mean()

plt.plot(gdppc["year"], gdppc["GDPPC, growth"], marker = 'o', color = '#9E3039')
plt.title('   ()         1993-2028.')
plt.axhline(y = average_growth, color = 'blue', label = "    ")
plt.xlabel('')
plt.ylabel('   ()    \n(%)')
plt.legend()
plt.grid(True)
```

```python
average_growth = gdppc["GDPPC, growth"].mean()

plt.plot(gdppc["GDPPC_log"], gdppc["GDPPC, growth"], marker = 'o', color = '#9E3039')
plt.title('   ()         1993-2028.')
plt.axhline(y = average_growth, color = 'blue', label = "    ")
plt.xlabel('log(   ()    )')
plt.ylabel('   ()    \n(%)')
plt.legend()
plt.grid(True)
```
